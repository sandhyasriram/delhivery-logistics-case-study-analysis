Creating new features

```python
trip['destination_name'] = trip['destination_name'].str.lower() #lowering all columns
trip['source_name'] = trip['source_name'].str.lower()
```

```python
def place2state(x):
  state = x.split('(')[1]
  return state[:-1] # removing from ending

def place2city(x):
  city = x.split('(')[0]
  city = city.split('_')[0]
  return city

def place2city_place(x):
# we'll remove states
  x = x.split('(')[0]

  len_ = len(x.split('_'))
  if len_ >= 3:
    return x.split('_')[1]

  # samll cities have same city and place name
  if len_ ==2:
    return x.split('_')[0]

  return x.split(' ')[0]


def place2code(x):
    # We will remove state
    x = x.split(' (')[0]

    if len(x.split('_')) >= 3 :
        return x.split('_')[-1]

    return 'none'
```

```python
trip['destination_state'] = trip['destination_name'].apply(lambda x: place2state(x))
trip['destination_city']  = trip['destination_name'].apply(lambda x: place2city(x))
trip['destination_place'] = trip['destination_name'].apply(lambda x: place2city_place(x))
trip['destination_code']  = trip['destination_name'].apply(lambda x: place2code(x))
```

```python
trip['source_state'] = trip['source_name'].apply(lambda x: place2state(x))
trip['source_city']  = trip['source_name'].apply(lambda x: place2city(x))
trip['source_place'] = trip['source_name'].apply(lambda x: place2city_place(x))
trip['source_code']  = trip['source_name'].apply(lambda x: place2code(x))
```

Hypothesis Teting

Compare the difference between od_time_diff_hour and start_scan_to_end_scan

```python
trip[['od_time_diff_hour', 'start_scan_to_end_scan']]
```

```python
sns.boxplot(data=trip[['od_time_diff_hour', 'start_scan_to_end_scan']])
plt.show()
```

Hypothesis Testing. H0: No significant difference between od_time_diff_hour and start_scan_to_end_scan.

Ha: There is a significance difference between od_time_diff_hour and start_scan_to_end_scan.

```python
from scipy.stats import ttest_ind

alpha = 0.05

tstat,p_value = ttest_ind(trip['start_scan_to_end_scan'], trip['od_time_diff_hour'],alternative='two-sided')
print('Tstatistcs:', tstat)
print('p_value:', p_value)

if p_value < alpha:
  print('Reject H0:There is a significance difference between od_time_diff_hour and start_scan_to_end_scan.')

else:
  print("Fail to reject H0: No significant difference between od_time_diff_hour and start_scan_to_end_scan.")
```

#Finding outliers in the numerical variables

```python
num_cols = trip.select_dtypes(include= ['int','float'])
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=num_cols, orient='h')
plt.xticks(rotation=90)
plt.title('Boxplot of Numerical columns')
plt.show()
```

```python
q1 = num_cols.quantile(0.25)
q3 = num_cols.quantile(0.75)

IQR = q3-q1

lower_bound = q1-1.5*IQR
upper_bound = q3+1.5*IQR
```

```python
outliers = ((num_cols < lower_bound)|(num_cols > upper_bound))
outlier_count = outliers.sum()
print(outlier_count)
```

```python
outlier_percentage = (outlier_count/len(num_cols))*100
outlier_percentage
```

```python
# filling the outliers with lower_bound and upper_bound.

trip_capped = trip.copy()
trip_capped[num_cols.columns] = num_cols.clip(lower=lower_bound, upper=upper_bound, axis=1)
```

```python
print('Before:', num_cols.min(), num_cols.max())
print('After:', trip_capped[num_cols.columns].min(), trip_capped[num_cols.columns].max())
```

```python
trip_capped[num_cols.columns].boxplot(figsize=(12,6))
plt.title("Boxplot After Capping Outliers")
plt.xticks(rotation=90)
plt.show()
```
